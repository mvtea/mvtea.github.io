First of all, in order to plot something, you need some data. Namely, an x-axis and a y-axis. This data, as you might guess, is stored in lists/arrays, so let's generate some fake data to use:

```python
import numpy as np
x = np.linspace(-100,100,100)
y = x**2
```

This should give us a parabola when we plot it, as we've just taken the square of all the value on our x-axis. Now, in order to do the actual plotting, we'll need another library called `matplotlib`, specifically the `pyplot` module. We import it like this, the common alias being `plt`:

```python
import matplotlib.pyplot as plt
```

Now we can get to work. All we need to do now is give the `matplotlib` function `plot()` our axes:

```python
plt.plot(x,y)
```

It's looking good so far, but it's missing labels, and a title. **Always title your plot and label your axes appropriately.** In order to add axis labels, we use the functions `xlabel()` and `ylabel()`, and to add a title, we use `title()`:

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
```

Now, perhaps this isn't the only function you want on your plot. For example, the infamous plot of the Planck function of different stellar temperatures has multiple blackbody curves on it. We can totally do that with `matplotlib`. Let's make a couple more fake functions to demonstrate this:

```python
y2 = 4*x**2
y3 = 10*x**2
```

In order to _overplot_ these functions on the plot we already have, we just add more `plot()` functions with these new axes:

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
```

Now, with all these lines, it might be hard to tell which correponds to which function. We can create a legend for the plot with the `legend()` function, but first we have to give each plot a _label_ (make sure to call `legend()` _after_ your plotting lines, or it'll have no labels to fill itself with):

```python
plt.plot(x,y, label='x^2')
plt.plot(x,y2, label='4x^2')
plt.plot(x,y3, label='10x^2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.legend()
```

(The legend goes wherever it wants; don't worry about formatting for now.) Now, just for the sake of demonstration, perhaps we want to have a scatter plot rather than a line plot. Instead of `plot()`, we can use `scatter()`:

```python
plt.scatter(x,y, label='x^2')
plt.scatter(x,y2, label='4x^2')
plt.scatter(x,y3, label='10x^2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.legend()
```

With a lot of points, this looks kind of silly, but you can tell that they're discrete values. Well great, you've got your plot. Let's say you're satisfied. If you want to save the plot, you can add the `savefig()` function to the end of your list of commands, which takes a string with the name you want to give the file as an argument. It'll save the figure in the same directory as your code by default:

```python
plt.scatter(x,y, label='x^2')
plt.scatter(x,y2, label='4x^2')
plt.scatter(x,y3, label='10x^2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.legend()
plt.savefig("test.pdf")
```

Note that the `savefig()` function overwrites any file with the name that you give it!

That's about it for the basics, but there are two more things I'll mention. First, just like printing, Jupyter likes to give you the output without you explicitly asking for it. The `print()` equivalent for `matplotlib` is the `show()` function:

```python
plt.scatter(x,y, label='x^2')
plt.scatter(x,y2, label='4x^2')
plt.scatter(x,y3, label='10x^2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.legend()
plt.show()
```

Note that you don't have to use the `show()` function if you're not planning on reading the plot to the screen, i.e. you just want to save it as a file. 

Second, once you're done with a plot, you should make it a good habit to use the `close()` function. When you use `show()` to read it to your screen, you do this implicitly by closing the plot window when you're done. In other cases, though, you should add the `close()` function to the end of your list of commands. Sometimes it will leave Python open and running, which can slow down your computer. Also, if you don't use `close()` after plotting, and go on to make another plot later in your code, you just might find that the results of the earlier plot show up on both, since you haven't told Python that you were done with the first one.

```python
plt.scatter(x,y, label='x^2')
plt.scatter(x,y2, label='4x^2')
plt.scatter(x,y3, label='10x^2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.legend()
plt.savefig("test.pdf")
plt.close()
```

Notice how Jupyter didn't show that one? It's because it knew that all I wanted to do was save the figure, which it figured out when it read `plt.close()`.
